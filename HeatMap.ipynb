```python
df1.corr()
```

```python
plt.figure(figsize=(16,9))
ax=sns.heatmap(df1.corr(), annot=True, linewidths=3)
plt.title("Who is responsible for Global Wrming", fontsize=30)
ax.tick_params(size=10, color='w', labelsize=10, labelcolor='w')
plt.show()
```
