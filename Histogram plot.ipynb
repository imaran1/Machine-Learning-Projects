# Histogram Plot

```python
import matplotlib.pyplot as plt
import numpy as np
import random
```

```python
ml_student_age=np.random.randint(18,45,(100))
py_student_age=np.random.randint(15,40,(100))
```

```python
print(ml_student_age)
print(py_student_age)
```

```python
plt.hist(ml_student_age)

plt.title("ML Student age Histogram Chart")
plt.xlabel("Student age cotegory")
plt.ylabel("No. Student age")
plt.show()
```

```python
bins=[15,20,25,30,35,35,40,45]
```

```python
plt.hist(ml_student_age, bins, rwidth=.8)

plt.title("ML Student age Histogram Chart")
plt.xlabel("Student age cotegory")
plt.ylabel("No. Student age")
plt.show()
```

```python
plt.hist(ml_student_age, bins, rwidth=.8, histtype="step")

plt.title("ML Student age Histogram Chart")
plt.xlabel("Student age cotegory")
plt.ylabel("No. Student age")
plt.show()

# 'bar', 'barstacked', 'step',  'stepfilled'
```

```python
plt.hist(ml_student_age, bins, rwidth=.8, histtype="bar", 
orientation='horizontal', color="m",label="ML Student")

plt.hist(py_student_age, bins, rwidth=.8, histtype="bar", 
orientation='horizontal', color="y",label="PY Student")

plt.title("ML Student age Histogram Chart")
plt.xlabel("Student age cotegory")
plt.ylabel("No. Student age")
plt.legend()
plt.show()
```

```python
plt.hist(ml_student_age, bins, rwidth=.8, histtype="bar", 
orientation='horizontal', color="m",label="ML Student")

plt.hist(py_student_age, bins, rwidth=.8, histtype="bar", 
orientation='vertical', color="y",label="PY Student")

plt.title("ML Student age Histogram Chart")
plt.xlabel("Student age cotegory")
plt.ylabel("No. Student age")
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(16,9))
plt.hist([ml_student_age, py_student_age], bins, rwidth=.8, histtype="bar", 
orientation='vertical', color=["m", "y"],
         label=["ML Student", "Py Student"])

# plt.hist(py_student_age, bins, rwidth=.8, histtype="bar", 
# orientation='vertical', color="y",label="PY Student")

plt.title("ML Student age Histogram Chart")
plt.xlabel("Student age cotegory")
plt.ylabel("No. Student age")
plt.legend()
plt.show()
```

```python
from matplotlib import style
style.use("ggplot")
plt.figure(figsize=(16,9))
plt.hist([ml_student_age, py_student_age], bins, rwidth=.8, histtype="bar", 
orientation='vertical', color=["m", "y"],
         label=["ML Student", "Py Student"])

plt.title("ML Student age Histogram Chart")
plt.xlabel("Student age cotegory")
plt.ylabel("No. Student age")
plt.legend()
plt.show()
```
