```python
df1=pd.read_csv("D:\\Machine-Learning-Practice\\My-Learning-Projects\\googleplaystore.csv", nrows=1000)
df1.shape
```

```python
df1
```

```python
x=df1["Rating"]
y=df1["Reviews"]
```

```python
plt.scatter(x,y)
```

```python
plt.scatter(x,y)
plt.title("Google Play Store Apps Scatter Plot Graph")
plt.xlabel("Rating")
plt.ylabel("Reviews")
```

```python
plt.figure(figsize=(16,9))
plt.scatter(x,y, c="g", s=170, alpha=.5)
plt.title("Google Play Store Apps Scatter Plot Graph")
plt.xlabel("Rating")
plt.ylabel("Reviews")
```

```python
plt.figure(figsize=(16,9))
plt.scatter(x,y, c="r", s=170, marker="*", alpha=.5, linewidths=8, edgecolor="g")
plt.title("Google Play Store Apps Scatter Plot Graph")
plt.xlabel("Rating")
plt.ylabel("Reviews")
```

```python
plt.figure(figsize=(16,9))
plt.scatter(x,y, c="g", s=170, marker="*",
            linewidths=8,)

plt.scatter(x,df1["Installs"], c="r", s=170, marker="o", 
  linewidths=8,)
# edgecolor="y"
# alpha=.5
# edgecolor="g"

plt.title("Google Play Store Apps Scatter Plot Graph")
plt.xlabel("Rating")
plt.ylabel("Reviews & Installs")
```
